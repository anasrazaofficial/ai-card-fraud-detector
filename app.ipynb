```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Load the dataset
creditCardData = pd.read_csv('./data/creditcard.csv')
```

```python
legit = creditCardData[creditCardData.Class == 0]
legit.shape
```

```python
fraud = creditCardData[creditCardData.Class == 1]
fraud.shape
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
creditCardData.groupby('Class').mean()
```

```python
legit_sample = legit.sample(n=492)
```

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.tail()
```

```python
new_dataset.Class.value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset.Class
```

```python
X
```

```python
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model = LogisticRegression(max_iter=10000)
```

```python
model.fit(X_train, Y_train)
```

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = str(round(accuracy_score(X_train_prediction, Y_train) * 100, 2)) + '%'
```

```python
training_data_accuracy
```

```python

X_test_prediction = model.predict(X_test)
test_data_accuracy = str(round(accuracy_score(X_test_prediction, Y_test) * 100, 2)) + '%'
```

```python
test_data_accuracy
```
